La solución de la ecuación 
$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2}$$
sujeta a las condiciones de frontera:
$$\frac{\partial u(0,t)}{\partial x}=0,\;\;\; \frac{\partial u(1,t)}{\partial x}=0, \;\;\;\;\; t>0$$
$$u(x,0)=f(x), \;\;\; 0<x<1$$
Para:
$$f(x)=\left\lbrace\begin{array}{c} -2 \;\; ~si\;\; 0<x\leq \frac{1}{2}\;\;\;\;  \\ 0 \; ~si \;\;\; \frac{1}{2}<x<1 \end{array}\right.$$

Esta dada en términos de la serie de Fourier por la expresión:

$$u(x,t)=-1 -\sum_{n=1}^{\infty} [\frac{4}{n\pi}\sin(\frac{n\pi}{2})] \cos(n\pi x)e^{-(n\pi)^2 t}.$$

Recordando que este problema modela la transferencia de calor en una barra ideal delgada, con  $k=1$ .

```python
from pylab import *
from matplotlib import animation,rc # biblioteca de animacion
from IPython.display import HTML # biblioteca para generar el video
%config InlineBackend.figure_format = 'retina'
```

Para empezar, graficamos la condición inicial.

```python
def fun(x):
  y =zeros(len(x))
  for i in range(len(x)):
    y[i]=-2
  return y
def gun(x):
  y =zeros(len(x))
  for i in range(len(x)):
    y[i]=0
  return y
```

```python
x= linspace(0,1,1000)
x1=[]
x2=[]
for i in range(len(x)):
  if(x[i]<=0.5):
    x1.append(x[i])
  else:
    x2.append(x[i])
f=fun(x1)
g=gun(x2)
plot(x1,f,label='$u(x,0)=-2$')
plot(x2,g,label='$u(x,0)=0$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Condiciones iniciales (u(x,0)=2)  (u(x,0)=0)")
xlim([-0.1,1.1]),ylim([-2.1,1.0]),grid(),legend()
show()
```

Aplicamos una Serie de Fourier para aproximar la condición inicial $u(x,t)=-2,   u(x,t)=0$

```python
def u_0(x,N=20): # N es el número de armónicos
    f = zeros(len(x))
    r = zeros(len(x))
    for n in range(1,N):
        f += ((4/n*pi)*sin(n*pi/2)*(cos(n*pi*x))) 
    r=-1-f
    return r
```

```python
f_n = u_0(x)
plot(x1,f,label='$u(x,0)=-2$')
plot(x2,g,label='$u(x,0)=0$')
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.1,1.1]),ylim([-11.5,11.5]),grid(),legend()
show()
```

Podemos visualizar la aproximación para diferentes cantidades de armónicos.

```python
plot(x1,f,label='$u(x,0)=-2$')
plot(x2,g,label='$u(x,0)=0$')
for i in range(1,22,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-11.5,11.5]),grid(),legend()
show()
```

Ahora, agregamos la parte temporal para visualizar la solución completa. Primero, para el momento inicial,  t=0 .

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    r = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += ((4/n*pi)*sin(n*pi/2)*(cos(n*pi*x)))*exp(-n**2*pi**2*t[i])
    r= -1-f
    return r
```

```python
xp = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)
```

```python
plot(x1,f,label='$u(x,0)=-2$')
plot(x2,g,label='$u(x,0)=0$')
plot(xp,U[:,0],label='$u(x,t)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$'); 
plt.title("Aproximación de la condición inicial")
xlim([-0.1,1.1]),ylim([-11.5,11.5]),grid(),legend()
show()
```

Luego, para diferentes tiempos, lo cual nos permite ver cómo la solución evoluciona cuando  $t→∞$ .

```python
plot(x,U[:,0],label='$u(x,t=0)$') # Mejorar
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,4],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=15)$')
plot(x,U[:,21],label='$u(x,t=21)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$'); 
plt.title("Evolución de la solución")
xlim([-0.1,1.1]),ylim([-11.5,11.5]),grid(),legend()
show()
```

Y podemos ver esto en una gráfica en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,1000)
t = linspace(0,1.0,1000)

X,T = np.meshgrid(x,t)

U = u_x_t(x,t)

plt.xlabel('$t$'); plt.ylabel('$x$'); 
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="plasma"); #Otros colores:"viridis", "inferno", "plasma"
```
